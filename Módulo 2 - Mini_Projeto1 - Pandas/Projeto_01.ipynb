## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
import missingno as msno
msno.matrix(cadastro_invest)
```

```python
quantidade = cadastro_invest.isnull().sum().sort_values(ascending=False)
porcentagem = (cadastro_invest.isnull().sum()/cadastro_invest.isnull().count()).sort_values(ascending=False)
dados_faltantes = pd.concat([quantidade, porcentagem], axis=1, keys=['Quantidade','Porcentagem'])
dados_faltantes
```

### Estratégia para tratar os dados faltantes:
1) No caso das colunas dos ativos, os dados faltantes NaN serão substituídos por 0 (zero), pois entendo que se não há valor é porque o investidor não aplicou no ativo em questão.  
2) No caso das notas, os dados faltantes NaN serão substituídos pela mediana (valor do meio das notas válidas), pois a ausência deste valor não significa que o cliente deu nota 0 (zero) e nem nota 100 (cem).


```python
values = {"Ativo01": 0, "Ativo01": 0, "Ativo02": 0, "Ativo03": 0, "Ativo04": 0, "Ativo05": 0, "Ativo06": 0,}
cadastro_invest = cadastro_invest.fillna(value=values)
```

```python
import missingno as msno
msno.matrix(cadastro_invest)
```

```python
cadastro_invest[['Nota1','Nota2']] = cadastro_invest[['Nota1','Nota2']].fillna(value=cadastro_invest[['Nota1','Nota2']].median())
cadastro_invest
```

## 7 - Manipule os dados a seguir:
* Criar coluna "Total" somando o valor de todos os ativos por investidor  
* Criar coluna "Media" com a média das notas 1 e 2  
* Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)  
* Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)  
* Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar a função map)  
* Transformar os dados da coluna Media em inteiro  
* Criar coluna "NPS" onde o investidor é classificado como promotor se a média das notas é maior ou igual a 80, como neutro se a média estiver entre 50 e 80, e como detrator se for abaixo de 50

```python
cadastro_invest['Total Ativos'] = cadastro_invest.iloc[:, 11:18].sum(axis=1)
```

```python
cadastro_invest['Média Notas'] = (cadastro_invest['Nota1'] + cadastro_invest['Nota2'])/2
```

```python
cadastro_invest['Domínio'] = cadastro_invest['Email'].str.split('@').str.get(1)
```

```python
cadastro_invest.rename(columns={'Salario':'Renda'}, inplace=True)
cadastro_invest['Renda'] = cadastro_invest['Renda'].str.split('$').str.get(1)
cadastro_invest['Renda'] = pd.to_numeric(cadastro_invest['Renda'])
```

```python
cadastro_invest['Sexo'] = cadastro_invest['Sexo'].map({'Masc':'Masculino', 'Fem':'Feminino'})
```

```python
cadastro_invest['Média Notas'] = cadastro_invest['Média Notas'].astype(int)
```

```python
cadastro_invest.loc[cadastro_invest['Média Notas'] >= 80, 'NPS'] = 'Promotor'
cadastro_invest.loc[(cadastro_invest['Média Notas'] < 80) & (cadastro_invest['Média Notas'] >= 50), 'NPS'] = 'Neutro'
cadastro_invest.loc[cadastro_invest['Média Notas'] < 50, 'NPS'] = 'Detrator'
```
